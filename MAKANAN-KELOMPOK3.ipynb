```python
import matplotlib.pyplot as plt
```

```python
dessert_sales = ['Lava_Cake', 'Mousse', 'Chocolate_Cake', 'Ice_Cream', 'Truffles', 'Brownie', 'Chocolate_Chip_Cookie', 'Chocolate_Pudding', 'Souffle', 'Chocolate_Cheesecake', 'Chocolate_Chips', 'Fudge', 'Mochi']
votes = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)
```

```python
plt.figure(figsize=(12, 6))
plt.barh(dessert_sales, votes)  # Membuat diagram batang horizontal dengan warna biru langit
plt.xlabel('Penjualan dalam Seminggu Terakhir')  # Label sumbu x
plt.title('Makanan')
```

```python
dessert_sales, votes = zip(*sorted(zip(dessert_sales, votes), key=lambda x: x[1], reverse=True)) #bagian zip sampai votes itu fungsinya untuk nyatuin daftar dessert dan votes, contoh : lava_Cake:14, mousse:15
plt.show()
```

```python
desserts_to_remove = dessert_sales[-3:]
print("Makanan Penutup yang dikeluarkan dari menu:", desserts_to_remove)
```

```python
chart yang cocok untuk menggambarkan kasus di atas adalah dengan menggunakan diagram batang horizontal, hal ini dilakukan karena :
    1. Memudahkan perbandingan panjang batang untuk menggambarkan popularitas relatif makanan.
    2. Mengurangi tumpang tindih antara label kategori, membuat visualisasi lebih mudah dibaca.
    3. Memungkinkan penempatan judul di atas diagram untuk memberikan gambaran umum tentang konten visualisasi.
```
